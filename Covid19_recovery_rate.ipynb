```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/web-data/data/cases_country.csv")
df.head()
```

```python
df.isna().sum()
```

<font color='blue'><h3> Dropping 'People_tested' and 'People_hospitalized' columns due to all NaN values present. </h3></font>

```python
df1 = df.drop(['People_Tested', 'People_Hospitalized'], axis=1)
```

```python
df1.shape
```

```python
df1.isna().sum()
```

```python
df1.corr()*100
```

<font color="blue"><h3> Dropping columns which do not have strong correlation and are not required. </h3></font>

```python
df2 = df1.drop(['Lat','Long_','UID','ISO3'], axis = 1)
```

```python
df2.corr()*100
```

```python
df2
```

<font color='blue'><h3> Adding new columns 'Unrecovered_Rate' and 'Recovery_Rate'. </h3></font>

```python
df2['Recovery_Rate'] = (df2['Recovered'] / df2['Confirmed'])*100

df2['Unrecovered_Rate'] = ((df2['Deaths'] + df2['Active'])/ df2['Confirmed'])*100
```

```python
df2
```

```python
df2.info()
```

<font color="blue"><h3> Replacing NA values with mean in the 'Recovered' column. </h3></font>

```python
df2['Recovery_Rate'] = df2['Recovery_Rate'].fillna(np.mean(df2.Recovery_Rate))
```

```python
#df2['Confirmed'] = df2['Confirmed'].astype(int)
#df2['Deaths'] = df2['Deaths'].astype(int)
#df2['Recovered'] = df2['Recovered'].astype(int)
#df2['Active'] = df2['Active'].astype(int)
#df2['Unrecovered'] = df2['Unrecovered'].astype(int)
```

```python
df2.info()
```

```python
df2
```

<font color='blue'><h3> Creating a DataFrame with the target and the independent variable. </h3></font>

```python
df3 = pd.DataFrame(data=df2, columns=['Unrecovered_Rate','Recovery_Rate'], index=np.arange(191))
df3
```

<font color='blue'><h3> Splitting data into train test split. </h3></font>

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

seed = 10
test_size = 0.30
training_Set, testing_Set = train_test_split(df3, test_size = test_size, random_state = seed)
train_x = training_Set[['Unrecovered_Rate']]
train_y = training_Set[['Recovery_Rate']]
test_x = testing_Set[['Unrecovered_Rate']]
test_y = testing_Set[['Recovery_Rate']]
```

```python
training_Set.shape , testing_Set.shape , df3.shape
```

<font color="blue"><h3> Fitting Linear Regression Model. </h3></font>

```python
#training the model
lm_1 = LinearRegression()
lm_1.fit(train_x, train_y)
```

```python
train_y_pred = lm_1.predict(train_x)

print("Train Set")
print("r2 score : ",r2_score(train_y, train_y_pred))
print("RMSE : ",mean_squared_error(train_y, train_y_pred, squared=False))
```

```python
test_y_pred = lm_1.predict(test_x)

print("Test Set")
print("r2 score : ",r2_score(test_y, test_y_pred))
print("RMSE : ",mean_squared_error(test_y, test_y_pred, squared=False))
```
